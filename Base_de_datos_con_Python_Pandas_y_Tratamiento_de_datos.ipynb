# **Selecciones y frecuencias**

-Seleccione solamente los inmuebles clasificados con tipo "Departamento"

-Seleccione los inmuebles clasificados con tipos 'Casa', 'Casa en condominio' y 'Casa de villa'

-Seleccione los inmuebles con area entre 60 y 100 metros cuadrados, incluyendo los limites

-Seleccione los inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,00

```python
import pandas as pd
```

```python
datos = pd.read_csv('Datos Residenciales.csv', sep=';')
datos.head(10)
```

```python
# -Seleccione solamente los inmuebles clasificados con tipo "Departamento"
seleccion = datos['Tipo'] == 'Departamento'
sec1 = datos[seleccion].shape[0]
sec1
```

```python
#-Seleccione los inmuebles clasificados con tipos 'Casa', 'Casa en condominio' y 'Casa de villa'
# El '|' es 'or'
seleccion = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa de villa')
sec2 = datos[seleccion].shape[0]
sec2
```

```python
#Seleccione los inmuebles con area entre 60 y 100 metros cuadrados, incluyendo los limites
#El '&' es el 'and'
seleccion = (datos['Area'] >= 60) & (datos['Area'] <= 100)
sec3 = datos[seleccion].shape[0]
sec3
```

```python
#Seleccione los inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,00
#Tenemos que combinar la columna 'Cuartos' y 'Valor'
seleccion = (datos['Cuartos'] >= 4) & (datos['Valor'] < 2000)
sec4 = datos[seleccion].shape[0]
sec4
```

# **Tratamiento de Datos Faltantes**
NaN

```python
import pandas as pd
```

```python
datos = pd.read_csv('Datos Residenciales.csv', sep=';')
datos.head(10)
```

```python
#como reconocer los datos NaN con isnull(), cuando sea nulo es true
datos.isnull()
```

```python
#Lo mismo pero al revez
datos.notnull()
```

```python
datos.info()
```

```python
#Se fija en valor si son nulos
datos['Valor'].isnull()
```

```python
#Pasa un dataframe con valores nulos o NaN
datos[datos['Valor'].isnull()]
```

```python
#dropna(subset = ['Valor']...) hace que borre los valores nulos
#A y B son el antes y despues de la cantidad de columnas, A con 9 y B vale 0
A = datos.shape[0]
datos.dropna(subset = ['Valor'], inplace = True)
B = datos.shape[0]
A - B
```

```python
#Como vemos, se fueron las 9 lineas nulas
datos[datos['Valor'].isnull()]
```

```python
datos
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
seleccion = (datos['Mantenimiento'].isnull()) & (datos['Tipo'] == 'Departamento')
```

```python
#Para seleccionar todo lo contrario a un criterio de seleccion se usa la ~
#Se eliminan 745 lineas
A = datos.shape[0]
datos = datos[~seleccion]
B = datos.shape[0]
A - B
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
#Rellenar todos los NaN con 0
datos = datos.fillna({'Mantenimiento':0,'Impuesto':0})
```

```python
datos[datos['Impuesto'].isnull()].shape[0]
```

# **Identificando y Excluyendo Outliers**

Utilizo Box=plot  o diagrama de caja
es un grafico estadistico que nos permite visualizar los datos

Valor-atipico.jpg




```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14,6))
```

```python
datos = pd.read_csv('alquiler_residencias_tratados2.csv', sep = ';')
datos.head(10)
```

```python
#Para usar boxplot, tiene que ser una columna/fila que se pueda medir
datos.boxplot(['Valor'])
```

```python
datos[datos['Valor'] >= 500000]
```

```python
valor = datos['Valor']
```

```python
#ESTO SE SACA DEL ANALISIS DE BOXPLOT
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1-1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
seleccion = (valor >= limite_inferior) & (valor <= limite_superior)
```

```python
datos_new = datos[seleccion]
```

```python
datos_new.boxplot(['Valor'])
```

```python
datos.hist(['Valor'])
```

```python
datos_new.hist(['Valor'])
```

Ahora por columna de clasificacion

```python
datos.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = datos.groupby('Tipo')['Valor']
```

```python
type(grupo_tipo)
```

```python
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1-1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
limite_inferior
```

```python
limite_superior
```

```python
datos_new2 = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = datos['Tipo'] == tipo
  eh_dentro_limite = ((datos['Valor'] >= limite_inferior[tipo]) & (datos['Valor'] <= limite_superior[tipo]))
  seleccion = eh_tipo & eh_dentro_limite
  datos_seleccion = datos[seleccion]
  datos_new2 = pd.concat([datos_new, datos_seleccion])
```

```python
datos_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
#Terminado esto lo comparto
datos_new.to_csv('alquileres_residenciales_sin_outliers.csv', sep = ';', index = False)
```
