# Traduciendo las variables

```python
datos.columns
```

```python
#Crear un diccionario para cambiar el nombre de las columnas
dic_colums = {'LIMIT_BAL':'Limite',
              'CHECKING_ACCOUNT':'CuentaCorriente',
              'EDUCATION': 'Educacion',
              'MARRIAGE':'EstadoCivil',
              'AGE':'Edad',
              'BILL_AMT': 'ValorFactura',
              'PAY_AMT':'ValorPago',
              'DEFAULT':'Mora'}
```

```python
#Creo otro dataframe con los nombres de las columnas cambiadas
tarjetas = datos.rename(columns = dic_colums)
tarjetas
```

```python
tarjetas.head(10)
```

```python
#Ahora le cambio el contenido a las columnas(las traduzco), utilizo unique para ver los valores unicos
tarjetas.CuentaCorriente.unique()
```

```python
#Hago el mismo proceso
dic_cuenta= {
    'Yes':'Si',
    'No':'No'
}
```

```python
#.Map() te permite transformar un iterable completo usando otra funciÃ³n, osea cambiar todo el array
tarjetas.CuentaCorriente.map(dic_cuenta)
```

```python
#Ahora lo cambio
tarjetas.CuentaCorriente = tarjetas.CuentaCorriente.map(dic_cuenta)
```

```python
tarjetas.head()
```

```python
#Ahora lo hacemos para la Educacion
tarjetas.Educacion.unique()
```

```python
dic_Educacion = {
    '2.University':'2.Universidad',
    '3.Graduate School':'3.GraduadoCole',
    '1.High School':'1.Colegio'
}
```

```python
tarjetas.Educacion = tarjetas.Educacion.map(dic_Educacion)
```

```python
tarjetas.head()
```

```python
tarjetas.EstadoCivil.unique()
```

```python
dic_EstadoCivil = {
    'Married':'Casado',
    'Single':'Soltero'
}
```

```python
tarjetas.EstadoCivil = tarjetas.EstadoCivil.map(dic_EstadoCivil)
```

```python
tarjetas.head()
```

# **Test de hipÃ³tesis**

utilizo la libreria

> https://docs.scipy.org/doc/scipy/reference/stats.html

Creamos una hipÃ³tesis

```python
#Utilizo la libreria scipy.stats
from scipy.stats import ranksums
```

H<sup>null<null>


---
La distribucion de los grupos con mora y sin mora son la misma


H<sup>null<null>


---
La distribucion de los grupos con mora y sin mora **NO** son la misma


```python
tarjetas.head()
```

```python
#Calcular cuantos son los que deben algo a partir de la mora
Deudor = tarjetas.query('Mora == 1').ValorFactura
Deudor
```

```python
NoDeudor = tarjetas.query('Mora == 0').ValorFactura
NoDeudor
```

```python
#Vemos la diferencia entre las 2
resultado = ranksums(Deudor, NoDeudor)
resultado
```

```python
print(f'El p-value de nuestro test es {resultado.pvalue}')
```
